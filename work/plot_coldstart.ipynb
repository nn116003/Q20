```python
# coding:utf-8
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = np.load(open("./coldstart_1029.npy",'rb'))
data_2 = np.load(open("./coldstart_long.npy",'rb'))
```

```python
df = pd.DataFrame({"random_top3":data[:5000],"top1":data_2[:5000]})
df['m3']  =np.convolve(df.random_top3, np.ones(100)/100., mode='same')
df['m1']  =np.convolve(df.top1, np.ones(100)/100., mode='same')
```

```python
#plt.plot(data[:5000])
x = np.arange(5000)
#plt.figure(figsize=(12, 20), dpi=200)
plt.title('accuracy of cold-start-games')
plt.plot(x, df.random_top3.values, 'b', alpha=0.3)
plt.plot(x, df.top1.values, 'g',alpha=0.3)
plt.plot(x[:4900], df.m3.values[:4900], 'b', label='random_top3')
plt.plot(x[:4900], df.m1.values[:4900], 'g', label='top1')
plt.xlabel('number of games')
plt.ylabel('accuracy')
plt.legend()
plt.savefig('csg.png',dpi=200)
```
